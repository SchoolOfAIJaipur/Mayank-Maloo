```python
import matplotlib.pyplot as plt
def pabEqn(a,x):
    return((x*x))
randomX=[]
randomY=[]
for i in range(-1000,1000):
    randomX.append(rand.randint(-1000,1000))
for i in range(0,len(randomX)):
    randomY.append(pabEqn(2,randomX[i]))
plt.plot(randomX,randomY)
plt.show()
```

```python
import matplotlib.pyplot as plt
def hyperbolaEqn(a,b,x):
    return((b/a)*((x*x-a*a)**0.5))
randomX=[]
randomY=[]
for i in range(-1000,1000):
    randomX.append(rand.randint(-1000,1000))
for i in range(0,len(randomX)):
    randomY.append(hyperbolaEqn(4,9,randomX[i]))
plt.plot(randomX,randomY)
plt.show()
    
```
